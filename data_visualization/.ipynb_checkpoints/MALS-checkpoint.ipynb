27 binnary categories -> 54 binary attribute <--> atribute prompts

```python
ATTRIBUTE_NAME = '''gender,age,length-hair,wearing-hat,carrying-backback,carrying-handbag,carrying-bag,
                  sleeve-length, length-of-lower-body, type-of-lower-body,
                  upper-body-black,upper-body-white,upper-body-red,upper-body-purple,
                  upper-body-yellow,upper-body-blue,upper-body-green,upper-body-gray,
                  lower-body-black, lower-body-white, lower-body-purple, lower-body-yellow,
                  lower-body-blue, lower-body-green, lower-body-pink, lower-body-gray,lower-body-brown
                  '''
ATTRIBUTE_NAME = re.sub(r"[\n\t\s]*", "", ATTRIBUTE_NAME).split(',')
```

```python
len(ATTRIBUTE_NAME)
```

```python
positive_attributes = ATTR_PROMPTS[1::2]
negative_attributes = ATTR_PROMPTS[0::2]
PROMPT_TABLE = pd.DataFrame({"attribute_category": ATTRIBUTE_NAME, "positive_attributes": positive_attributes, "negative_attributes": negative_attributes})
PROMPT_TABLE
```

## View image - captions in table

```python
df = pd.DataFrame(ANNO)
df.head(1)
```

```python
def get_attribute_meaning(annot):
    labels = PROMPT_TABLE['attribute_category']
    meaning = np.array([""]*27, dtype='<U50')
    annot = np.array(annot)
    meaning[annot==1] = PROMPT_TABLE['positive_attributes'][annot==1]
    meaning[annot==0] = PROMPT_TABLE['negative_attributes'][annot==0]
    for i, prompt in enumerate(meaning):
        for re_word in ["the", "person", "does", "is"]:
            meaning[i] = meaning[i].replace(re_word, '')
    return str(meaning[meaning!=""]).replace('[', '').replace(']', '').replace('  ', '<br><br>')
```

```python
df['image'] = df['image'].apply(lambda x: os.path.join(IMAGE_PATH, x))
df['label'] = df['label'].apply(get_attribute_meaning)
if 'image_id' in df: df.drop(columns=["image_id"], inplace=True)
```

# Interactive GUI

```python
def get_attribute_meaning(annot):
    # labels = ['gender','age','hair','hat','backpack','handbag','bag','up','down','clothes',
    #           'upblack','upwhite','upred','uppurple','upyellow','upblue','upgreen','upgray',
    #           'downblack','downwhite','downpurple','downyellow','downblue','downgreen', 'downpink','downgray','downbrown']
    labels = PROMPT_TABLE['attribute_category']
    meaning = np.array([""]*27, dtype='<U50')
    annot = np.array(annot)
    meaning[annot==1] = PROMPT_TABLE['positive_attributes'][annot==1]
    meaning[annot==0] = PROMPT_TABLE['negative_attributes'][annot==0]
    for i, prompt in enumerate(meaning):
        for re_word in ["the", "person", "does", "is"]:
            meaning[i] = meaning[i].replace(re_word, '')
    # meaning[annot==-1] = "can't be infered"
    return pd.DataFrame({
        # 'Label': labels[meaning!=""], 
        # 'Value': annot[meaning!=""], 
        
        'Prompt': meaning[meaning!=""]})
```

```python
import random, time
import pandas as pd
def generate_random():
    k = random.randint(0, len(ANNO))
    demo_text = ANNO[k]['caption']
    demo_img = PIL.Image.open(IMAGE_PATH+ANNO[k]['image']).convert('RGB')
    demo_attr = get_attribute_meaning(ANNO[k]['label'])
    return demo_text, demo_img, demo_attr
```

```python
with gr.Blocks() as demo:
    with gr.Row():
        summit_button_0 = gr.Button(value="Click to show image")
    for i in range(3):
        demo_text, demo_img, demo_attr = generate_random()
        with gr.Row():
            random_text = gr.Textbox(label="Caption of the image", value = demo_text)
        with gr.Row():
            with gr.Column():
                with gr.Row():
                    random_image = gr.Image(type="pil", value = demo_img)
            with gr.Column():
                random_attr = gr.Dataframe(label="Attribute prompts", value = demo_attr)
    summit_button_0.click(, inputs=None, outputs=[random_text, random_image, random_attr])
demo.launch(inline=True, share=True)
```
