```python
import os, sys
from pathlib import Path
```

```python
# get all file .ipynb
ipynb_files = list(Path('.').glob('*.ipynb'))
```

```python
ipynb_files = sorted([str(f) for f in ipynb_files])
```

```python
ipynb_files = ipynb_files[:-1]
```

```python
ipynb_files
```

```python
import jupyter
for f in ipynb_files: 
    file_save = 'docs/' + f[:-6] + 'md'
    print(file_save)
    os.system('jupyter nbconvert --to markdown ' + f + ' --output ' + file_save)
```
